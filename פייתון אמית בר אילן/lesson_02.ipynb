### Part 2: Checking Variable Types
Use the type() function to check what type a variable is:  
Code cell:

```python
# Let's check different types
age = 15
height = 1.75
name = "David"
is_student = True

print(type(age))           # <class 'int'>
print(type(height))        # <class 'float'>
print(type(name))          # <class 'str'>
print(type(is_student))    # <class 'bool'>
```

---
### Part 3: Type Conversion
Sometimes we need to convert between types:  
Converting to Integer

```python
text = "25"
number = int(text)
print(number + 5)  # Result: 30
```

Converting to Float


```python
text = "3.14"
number = float(text)
print(number * 2)  # Result: 6.28
```

Converting to String


```python
age = 15
text = str(age)
print("I am " + text + " years old")
```

Important! You can't convert text that isn't a number:


```python
int("hello")  # This will cause an ERROR!
```

### Part 4: Working with User Input
The input() function lets us get information from the user:

```python
name = input("What is your name? ")
print(f"Hello {name}!")
```

Important: input() always returns a string, even if the user types a number!

```python
# Wrong way:
age = input("How old are you? ")
future = age + 10  # ERROR! Can't add number to string

# Correct way:
age = int(input("How old are you? "))
future = age + 10  # This works!
```

### Task 4: Division Practice
Write code that divides 17 by 5 and shows:

1. Regular division (with decimal)
2. Integer division (whole number only)
3. The remainder (modulo)

Code cell:

```python
# Regular division
# Numbers
a = 17
b = 5

# 1. Regular division
regular_div = a / b
print("Regular division:", regular_div)  # 3.4

# 2. Integer division
int_div = a // b
print("Integer division:", int_div)      # 3

# 3. Remainder (modulo)
remainder = a % b
print("Remainder:", remainder)           # 2
```

2. Time Converter:

- Calculate circumference: 2 × π × radius (use 3.14159 for π)
- Calculate area: π × radius²
- Print both results


4. Savings Calculator:

- Ask how much money you save per week
- Calculate savings after 1 month, 6 months, 1 year
- Calculate if you add 5% interest per year



Code cell for homework:

```python
# Space for your homework
# Ask for weight and height
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# Calculate BMI
bmi = weight / (height ** 2)

# Print result
print(f"Your BMI is: {bmi:.2f}")
```

```python
# Ask for number of hours
hours = float(input("Enter number of hours: "))

# Convert
minutes = hours * 60
seconds = hours * 3600
days = hours / 24

# Print results
print(f"{hours} hours is equal to {minutes} minutes, {seconds} seconds, and {days:.2f} days.")
```

```python
# Ask for radius
radius = float(input("Enter the radius of the circle: "))
pi = 3.14159

# Calculate circumference and area
circumference = 2 * pi * radius
area = pi * (radius ** 2)

# Print results
print(f"Circumference: {circumference:.2f}")
print(f"Area: {area:.2f}")
```

```python
# Ask how much money you save per week
weekly_saving = float(input("How much do you save per week? "))

# Calculate savings
one_month = weekly_saving * 4
six_months = weekly_saving * 26
one_year = weekly_saving * 52

# Add 5% interest per year
one_year_with_interest = one_year * 1.05

# Print results
print(f"After 1 month: {one_month}")
print(f"After 6 months: {six_months}")
print(f"After 1 year: {one_year}")
print(f"After 1 year with 5% interest: {one_year_with_interest:.2f}")
```
